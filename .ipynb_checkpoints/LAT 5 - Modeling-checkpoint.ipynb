## Load dataset

```python
iris = datasets.load_iris()
df_data = pd.DataFrame(data=np.c_[
                       iris['data'], iris['target']], columns=iris['feature_names'] + ['target'])
print(df_data.shape[0])
df_data.head()
```

## Numerical summary

```python
df_data.describe()
```

```python
df_data['target'].value_counts(dropna=False)
```

## Pre-modelling

### Exploratory data analysis

```python
def compare_continous(df, col, target):
    plt.figure(figsize=(15,10))
    target_values = set(df[target])
    sns.violinplot(x=target, y=col, data=df)
    plt.show()
```

```python
target_cols = ['sepal length (cm)', 'sepal width (cm)',
               'petal length (cm)', 'petal width (cm)']
for c in target_cols:
    print(c)
    compare_continous(df_data, c, 'target')
    print('='*105)
```

```python
df_data.corr(method='pearson').style.format("{:.2}").background_gradient(cmap=plt
                                                                         .get_cmap('coolwarm'), axis=1)
```

```python
def plot_parallel_corrdinates(df, target): 
    plt.figure(figsize=(15,10))
    parallel_coordinates(df, target)
    plt.show()
```

```python
plot_parallel_corrdinates(df_data, 'target')
```

### Train-test split

```python
train, test = train_test_split(
    df_data, test_size=0.25, stratify=df_data['target'], random_state=42)
features = ['sepal length (cm)', 'sepal width (cm)',
            'petal length (cm)', 'petal width (cm)']
X_train = train[features]
y_train = train['target']
X_test = test[features]
y_test = test['target']
```

## Modelling

```python
def plot_cm(model_name, model, X_test, y_test):
    fig, ax = plt.subplots(figsize=(15, 10))
    plot_confusion_matrix(model,
                          X_test,
                          y_test,
                          display_labels=set(df_data['target']),
                          cmap=plt.cm.Blues,
                          normalize=None,
                          ax=ax)
    ax.set_title('{} Confusion Matrix'.format(model_name))
    plt.show()
    print('='*110)
```

```python
def run_model(model_name, model, X_train, y_train, X_test, y_test):
    model.fit(X_train, y_train)
    prediction = model.predict(X_test)
    accuracy = accuracy_score(prediction, y_test)
    f1 = f1_score(prediction, y_test, average='macro')
    precision = precision_score(prediction, y_test, average='macro')
    recall = recall_score(prediction, y_test, average='macro')
    plot_cm(model_name, model, X_test, y_test)
    return model, accuracy, f1, precision, recall
```

```python
def plot_tree_model(model):
    plt.figure(figsize=(15, 10))
    plot_tree(model, feature_names=features,
              class_names=[str(d) for d in list(set(df_data['target']))], filled=True)
    plt.show()
```

```python
model_names = ['Decision Tree', 'Naive Bayes', 'kNN', 'SVM', 'Logistic Regression',
               'Random Forest', 'Neural Network', 'Gradient Boosting', 'AdaBoost']
df_model_performance = pd.DataFrame(columns=[
                                    'model', 'accuracy', 'macro_f1_score', 'macro_precision_score', 'macro_recall_score'])
for mn in model_names:
    print('PROCESSING {} MODEL'.format(mn.upper()))
    if mn == 'Decision Tree':
        model = DecisionTreeClassifier(random_state=42)
    elif mn == 'Naive Bayes':
        model = GaussianNB()
    elif mn == 'kNN':
        model = KNeighborsClassifier()
    elif mn == 'SVM':
        model = SVC(random_state=42)
    elif mn == 'Logistic Regression':
        model = LogisticRegression(random_state=42)
    elif mn == 'Random Forest':
        model = RandomForestClassifier(random_state=42)
    elif mn == 'Neural Network':
        model = MLPClassifier(random_state=42)
    elif mn == 'Gradient Boosting':
        model = GradientBoostingClassifier(random_state=42)
    elif mn == 'AdaBoost':
        model = AdaBoostClassifier(random_state=42)

    fitted_model, accuracy, f1, precision, recall = run_model(
        mn, model, X_train, y_train, X_test, y_test)

    df_model_performance = df_model_performance.append({
        'model': mn,
        'accuracy': accuracy,
        'macro_f1_score': f1,
        'macro_precision_score': precision,
        'macro_recall_score': recall
    }, ignore_index=True)

    if mn in ['Decision Tree', 'Random Forest', 'Gradient Boosting', 'AdaBoost']:
        print('Feature importance of {}:'.format(mn))
        for feature, importance in zip(features, fitted_model.feature_importances_):
            print(feature + ' importance score = ' + str(importance))
        print('='*110)
    if mn in ['Decision Tree']:
        print('{} Visualization'.format(mn))
        plot_tree_model(fitted_model)
    print('*'*110)
```

```python
df_model_performance
```

# Clustering
Dataset: https://www.kaggle.com/vjchoudhary7/customer-segmentation-tutorial-in-python
1. K-Means

## Load dataset

```python
df_data = pd.read_csv('Mall_Customers.csv')
print(df_data.shape[0])
df_data.head()
```

```python
# encode atribut kategorikal (Gender): 'M'->'1' dan 'F'->'0'
df_data['Gender'] = df_data['Gender'].astype('category')
df_data['Gender'] = df_data['Gender'].cat.codes
df_data.head()
```

## Numerical summary

```python
df_data.describe()
```

## Pre-modelling

### Standarization

```python
# normalisasi data di setiap kolom, kecuali CustomerID
df_temp = df_data[list(set(df_data.columns)-set(['CustomerID']))].apply(zscore)
df_data_norm = df_data[['CustomerID']].merge(df_temp, left_index=True, right_index=True)
df_data_norm.head()
```

```python
points = df_data_norm[list(set(df_data_norm.columns)-set(['CustomerID']))].values
```

## Modelling

```python
n_clusters = 4
model = KMeans(n_clusters=n_clusters,
               random_state=42)
model.fit(points)
df_data['Cluster'] = model.labels_
```

```python
def visualize_cat(data, column_name):
    fig, axes = plt.subplots(figsize=(15, 5), nrows=1, ncols=n_clusters)
    for c in set(model.labels_):
        df_cluster = data[data['Cluster'] == c]
        count = df_cluster[column_name].value_counts(normalize=True)
        axes[c].bar(count.index, count.values)
        axes[c].text(0.5, -0.2, 'Cluster ' + str((c)),
                     horizontalalignment='center', transform=axes[c].transAxes)
    plt.show()


def visualize_num(data, column_name):
    plt.figure(figsize=(15,10))
    sns.violinplot(x='Cluster', y=column_name, data=data)
    plt.show()
```

## Exploratory data analysis

```python
print('Jumlah pelanggan di setiap klaster:')
print(df_data.groupby('Cluster')['CustomerID'].nunique())
print('='*110 + '\n')

print('Distribusi gender di setiap klaster:')
visualize_cat(df_data, 'Gender')
print('='*110 + '\n')

for c in set(df_data.columns) - set(['CustomerID', 'Gender', 'Cluster']):
    print('Distribusi {} di setiap klaster:'.format(c))
    visualize_num(df_data, c)
    print('='*110 + '\n')
```
