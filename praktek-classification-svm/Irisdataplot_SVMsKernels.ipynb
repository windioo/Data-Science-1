```python
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC
import pandas as pd
# Loading some example data
iris = datasets.load_iris()
```

```python
# Import data from the external CSV
irisdata = pd.read_csv('iris.csv')
```

```python
irisdata.head()
```

```python
# select two columns 
X=irisdata[['SepalLengthCm', 'SepalWidthCm']].to_numpy() 
y=irisdata[['Species']].to_numpy()
```

```python
X = iris.data[:, :2] # we only take the first two features.
y = iris.target
```

```python
# Training a classifier
svm = SVC(C=10, kernel='linear',decision_function_shape='ovo')
svm.fit(X, y)

# Plotting decision regions
plot_decision_regions(X, y, clf=svm, legend=2)

# Adding axes annotations
plt.xlabel('sepal length [cm]')
plt.ylabel('petal length [cm]')
plt.title('SVM on Iris_linear')
plt.show()
```

```python
# Training a classifier
svm = SVC(C=0.000000001, kernel='linear',decision_function_shape='ovo')
svm.fit(X, y)

# Plotting decision regions
plot_decision_regions(X, y, clf=svm, legend=2)

# Adding axes annotations
plt.xlabel('sepal length [cm]')
plt.ylabel('petal length [cm]')
plt.title('SVM on Iris_linear')
plt.show()
```

```python
# Training a classifier
svm = SVC(C=1, kernel='rbf', gamma=15)
svm.fit(X, y)

# Plotting decision regions
plot_decision_regions(X, y, clf=svm, legend=2)

# Adding axes annotations
plt.xlabel('sepal length [cm]')
plt.ylabel('petal length [cm]')
plt.title('SVM on Iris_rbf')
plt.show()
```

```python
# Training a classifier
svm = SVC(C=1, kernel='rbf', gamma=0.5)
svm.fit(X, y)

# Plotting decision regions
plot_decision_regions(X, y, clf=svm, legend=2)

# Adding axes annotations
plt.xlabel('sepal length [cm]')
plt.ylabel('petal length [cm]')
plt.title('SVM on Iris_rbf')
plt.show()
```

```python
# Training a classifier
svm = SVC(C=0.1, kernel='sigmoid')
svm.fit(X, y)

# Plotting decision regions
plot_decision_regions(X, y, clf=svm, legend=2)

# Adding axes annotations
plt.xlabel('sepal length [cm]')
plt.ylabel('petal length [cm]')
plt.title('SVM on Iris_sigmiod')
plt.show()
```

```python
# Training a classifier
svm = SVC(C=10, kernel='poly', gamma=0.004)
svm.fit(X, y)

# Plotting decision regions
plot_decision_regions(X, y, clf=svm, legend=2)

# Adding axes annotations
plt.xlabel('sepal length [cm]')
plt.ylabel('petal length [cm]')
plt.title('SVM on Iris_Polynomial')
plt.show()
```

```python

from sklearn.model_selection import GridSearchCV 
  
# defining parameter range 
param_grid = {'C': [0.1, 1, 10, 100, 1000],  
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001], 
              'kernel': ['rbf','linear','poly']}  
  
grid = GridSearchCV(SVC(), param_grid, refit = True) 
  
# fitting the model for grid search 
grid.fit(X, y) 
print(grid.best_params_)
print(grid.best_estimator_) 
```
