```python
df_can=pd.read_excel('UN_MigFlow_All_CountryFiles\Canada.xlsx',
                        sheet_name='Canada by Citizenship',
                        skiprows=20,
                        skipfooter=1)
df_can
```

```python
df_can = df_can.set_index('OdName')
```

```python
df_can
```

```python
df_can= df_can.drop(columns=['Type','Coverage','AREA','REG',"DEV"])
```

```python
df_can
```

```python
years = list(map(str, range(1980,2014)))
df_haiti = df_can.loc[['Haiti']]
df_haiti = df_haiti.drop(columns=['AreaName','RegName','DevName'])
df_haiti = df_haiti.transpose()
df_haiti.plot(kind='line')
plt.title('Imigration from Haiti to China')
plt.ylabel('Number of Imigrations')
plt.xlabel('Years')
plt.show
```

```python
df_can2 = df_can.drop(columns=['AreaName','RegName','DevName'])
df_can.loc[:, 'Total']= df_can2.sum(axis=1)
df_can2 = df_can2.sort_values(['Total'], ascending=False, axis=0)
```

```python
df_can2
```

```python
df_top5 = df_can2.head()
years = list(range(1980,2014))
df_top5.loc[:, years].transpose()
```

```python
df_top5.plot(kind='area')
```

```python
count, bin_edges = np.histogram(df_can[2013])
df_can[2013].plot(kind='hist', xticks=bin_edges)
plt.ylabel('Number of Countries')
plt.xlabel('Numbers of Imigrants')
plt.show()
```

```python
df_can[2013].plot(kind='bar')
plt.ylabel('Number of Imigrations from Countries')
plt.xlabel('Countries')
plt.show()
```

```python
df_can['Total'] = df_can.loc[:,list(range(1980,2014))].sum(axis=1)
df_continents = df_can.groupby('AreaName', axis=0).sum()
df_continents
```

```python
df_continents['Total'].plot(kind='pie',autopct='%1.1f%%')
plt.title('Imigrations to Canada by Continents (1980-2013)')
plt.show()
```

```python
df_china = df_can.loc[['China'], list(range(1980,2014))].transpose()
df_china
```

```python
df_china.plot(kind='box',showmeans=True)
plt.title('Box plot of Chinese imigrants from 1980 - 2013')
plt.ylabel('Number of Imigrants')
plt.rcParams['figure.figsize'] = [30, 5]
plt.show()
```

```python
df_can.loc['Total',:] = df_can.sum(axis=0)
df_total = df_can.loc[['Total'], list(range(1980,2014))]
df_total = df_total.rename_axis('Total')
df_total = df_total.rename_axis('Year',axis=1)
df_total = df_total.transpose()
df_total = df_total.reset_index()
df_total
df_total.plot(kind='scatter', x='Year', y='Total')
```
